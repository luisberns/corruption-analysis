```python
for c in df.groupby('country').groups:
    x=df[df['country'] == c]
    x.plot('year', x.columns[2:].to_series(), title=c, figsize=(10,5))
```

```python
for c in df.groupby('country').groups:
    x=df[df['country'] == c]
    print('country: ', c, '\n', x.corr(), '\n')
```

```python
df_in = df.set_index(['country', 'year'])
result_list = []

for c in list(df.groupby('country').groups):
    country = df_in.loc[c]
    year_min = country.index.min()
    year_max = country.index.max()
    
    result = ((country.loc[year_max]/country.loc[year_min])-1)*100
    
    result_list.append([c] + list(result))
    
df_out = pd.DataFrame(result_list, columns=['country', 'cpi', 'internet_users', 'journalists_killed'])    
```

```python
# Here we can analyze the countries that improve to the contries that lowered their CPI comparing 2012 to 2016
df_out.set_index('country').sort_values('cpi', ascending=False)['cpi']
```
